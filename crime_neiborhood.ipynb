```python
import gmaps
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Google developer API key
from config import g_key
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
 # Create aiport dataframe
crime_df = pd.read_csv('Resources/crime_09_20_clean.csv')
crime_df.dropna(inplace = True) 
crime_df.head()
```

```python
crime_2018=crime_df.loc[crime_df['occur_year']==2018,:]
#crime_2018.head(5)
crime_2019=crime_df.loc[crime_df['occur_year']==2019,:]
crime_2020=crime_df.loc[crime_df['occur_year']==2020,:]
```

```python
grouped_neighbor_2018=crime_2018.groupby(['neighborhood'])
print(grouped_neighbor_2018)
grouped_2018=grouped_neighbor_2018.count()
sort_2018=grouped_2018.sort_values(['offense_id'],ascending=False)
sort_2018_df=pd.DataFrame(sort_2018['offense_id'])
sort_2018_df=sort_2018_df.rename(columns={"offense_id": "2018"})
sort_2018_df.head(20)
```

```python
grouped_neighbor_2019=crime_2019.groupby(['neighborhood'])
print(grouped_neighbor_2019)
grouped_2019=grouped_neighbor_2019.count()
sort_2019=grouped_2019.sort_values(['offense_id'],ascending=False)
sort_2019_df=pd.DataFrame(sort_2019['offense_id'])
sort_2019_df=sort_2019_df.rename(columns={"offense_id": "2019"})
sort_2019_df.head(20)
```

```python
grouped_neighbor_2020=crime_2020.groupby(['neighborhood'])
print(grouped_neighbor_2020)
grouped_2020=grouped_neighbor_2020.count()
sort_2020=grouped_2020.sort_values(['offense_id'],ascending=False)
sort_2020_df=pd.DataFrame(sort_2020['offense_id'])
sort_2020_df=sort_2020_df.rename(columns={"offense_id": "2020"})
sort_2020_df.head(20)
```

```python
crime2018_grouped=crime_2018.groupby(['neighborhood','lat','long'],as_index=False).count()
crime2018_grouped_df=pd.DataFrame(crime2018_grouped[['neighborhood',"lat","long","offense_id"]])
crime2018_grouped_df=crime2018_grouped_df.dropna(how="any")
crime2018_grouped_df=crime2018_grouped_df.rename(columns={"offense_id": "2018"})
crime2018_grouped_df
```

```python
crime2019_grouped=crime_2019.groupby(['neighborhood','lat','long'],as_index=False).count()
crime2019_grouped_df=pd.DataFrame(crime2019_grouped[['neighborhood',"lat","long","offense_id"]])
crime2019_grouped_df=crime2019_grouped_df.dropna(how="any")
crime2019_grouped_df=crime2019_grouped_df.rename(columns={"offense_id": "2019"})
crime2019_grouped_df
```

```python
crime2019_grouped=crime_2019.groupby(['neighborhood'],as_index=False).mean()
crime2019_grouped_df=pd.DataFrame(crime2019_grouped[['neighborhood',"lat","long","offense_id"]])
crime2019_grouped_df=crime2019_grouped_df.dropna(how="any")
merge_2019=pd.merge(sort_2019_df,crime2019_grouped_df,on ="neighborhood", how="left")
crime2019_grouped_df=crime2019_grouped_df.rename(columns={"offense_id": "2019"})
merge_2019_10=merge_2019.head(10)
merge_2019
```

```python
crime2020_grouped=crime_2020.groupby(['neighborhood'],as_index=False).mean()
crime2020_grouped_df=pd.DataFrame(crime2020_grouped[['neighborhood',"lat","long","offense_id"]])
crime2020_grouped_df=crime2020_grouped_df.dropna(how="any")
merge_2020=pd.merge(sort_2020_df,crime2020_grouped_df,on ="neighborhood", how="left")
crime2020_grouped_df=crime2020_grouped_df.rename(columns={"offense_id": "2020"})
merge_2020_10=merge_2020.head(10)
merge_2020
```
